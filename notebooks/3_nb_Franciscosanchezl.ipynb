```python
df_combinado=pd.concat([df2019,df2020])
df_sii = pd.melt(df_combinado,id_vars=["year",'Día'],value_vars=meses,var_name='Mes')
df_sii=df_sii[["year","Mes","Día","value"]]
df_sii.head()
df_sii.to_csv('C:/Users/franc/Documents/Magister Data science/Segundo trimestre/Analisis de datos/Tarea/tarea1-add/data/interim/df_sii.csv',
                  sep=';',index=False)
```

```python
print(df_sii.groupby(by=["Mes"],as_index = True)["value"].agg("mean"))
print("\n")
print(df_sii.groupby(by=["year","Mes"],as_index = False)["value"].agg("mean"))

```

Index = False elimina el índice de la fila, el agrupar por mes, dará como resultado el valor agrupodo incluyendo los años, mientras que al dejarlo por año-mes, separa estos dos y agrupa cada mes de cada año de manera distinta.
